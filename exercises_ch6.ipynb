## 演習2

```python
prices = [12, 15, 13, 18, 20]
```

### (1)

```python
pindex = []    # 価格指数を格納するリスト

for p in prices:

    idx = 100 * p / 13
    pindex.append(idx)

pindex
```

### (2)

```python
# 方法1

inflation = []

for p0, p1 in zip(prices[:-1], prices[1:]):

    res = p1/p0-1
    inflation.append(res)

inflation
```

次の方法もあるが，可読性と保守性が低いのが欠点となる。方法1が推奨される。

```python
# 方法2

inflation = []

for i in range( len(prices)-1 ):

    res = prices[i+1]/prices[i]-1
    inflation.append(res)

inflation
```

## 演習4

### (1)

```python
year = 0
ngdp = 591_345.7
debt = 1420_415.6

g = 0.02

while 100 * debt/ngdp > 100:

    ngdp *= (1+g)
    year += 1

year
```

### (2)

```python
year = 0
ngdp = 591_345.7
debt = 1420_415.6

g = 0.02
d = 0.01

while 100 * debt/ngdp > 100:

    debt *= (1+d)
    ngdp *= (1+g)
    year += 1

year
```

### (3)

```python
def years_to_achieve_target(g, d, target):

    year = 0
    ngdp = 591_345.7
    debt = 1420_415.6

    while 100 * debt/ngdp > target:

        debt *= ( 1+d )
        ngdp *= ( 1+g )
        year += 1

    return year

years_to_achieve_target(0.02, 0.01, 50)
```

下のコードは`g<d`の場合であり，理論的には政府負債対名目GDP比は発散するため，無限ループに入るはずです。しかし，この場合，Pythonの浮動小数点演算の限界により`25172`と表示されています。補足すると，Pythonは15～17桁の精度がありますが，非常に大きい，または，非常に小さい数の割り算などで誤差が蓄積し，演算結果が理論上の真値とズレることがあります。

```python
years_to_achieve_target(0.02, 0.03, 50)
```

改良版として次のコードを書くことができます。

```python
def years_to_achieve_target_revised_1(g, d, target):
    year = 0
    ngdp = 591_345.7
    debt = 1420_415.6
    prev_ratio = 100 * debt / ngdp

    if prev_ratio <= target:
        return "targetはすでに達成されています。"

    elif g <= d:
        return "targetを達成することは不可能"

    while 100 * debt/ngdp > target:

        debt *= ( 1+d )
        ngdp *= ( 1+g )
        year += 1

    return year
```

```python
years_to_achieve_target_revised_1(0.05, 0.04, 50)
```

```python
years_to_achieve_target_revised_1(0.05, 0.05, 350)
```

```python
years_to_achieve_target_revised_1(0.03, 0.05, 50)
```

また，`while`ループを次のように修正する方が可読性が高いかもしれません。

```python
def years_to_achieve_target_revised_2(g, d, target):
    year = 0
    ngdp = 591_345.7
    debt = 1420_415.6
    prev_ratio = 100 * debt / ngdp

    if prev_ratio <= target:
        return "targetはすでに達成されています。"

    elif g <= d:
        return "targetを達成することは不可能"

    while True:

        debt *= (1 + d)
        ngdp *= (1 + g)
        year += 1
        ratio = 100 * debt / ngdp

        if ratio <= target:
            return year
```

```python
years_to_achieve_target_revised_2(0.05, 0.04, 50)
```

```python
years_to_achieve_target_revised_2(0.05, 0.05, 350)
```

```python
years_to_achieve_target_revised_2(0.03, 0.05, 50)
```
