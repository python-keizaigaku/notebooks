### (3)

```python
random.seed(123)

def generate_rv(n):

    rv = [random.gauss() for _ in range(n)]
    return rv

generate_rv(5)

# 関数から関数の外にある.seed(123)にアクセスしている。
```

### (4)

```python
def calculate_mean(n, seed=None):

    random.seed(seed)
    rv = [random.gauss() for _ in range(n)]
    avr = sum(rv) / n
    return avr

calculate_mean(100, 123)
```

### (5)

```python
for _ in range(5):
    random.seed(123)
    print( random.gauss() )
```

`for`ループは実行箇所を毎回上から実行するため，同じ値が表示されています。`random.seed(123)`を`for`ループの上に移動して実行してみよう。

```python
random.seed(123)

for _ in range(5):
    print( random.gauss() )
```
